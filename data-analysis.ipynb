# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 50)
y = np.cos(x)
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 50)
y = x**2
plt.plot(x, y)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x, x**2, label='no')
plt.plot(x, x**3, label='not')
plt.plot(x, x**2 + 1.5 -x, label='n')
plt.title('test')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
